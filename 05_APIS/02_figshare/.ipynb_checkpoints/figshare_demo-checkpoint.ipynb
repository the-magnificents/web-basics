# Consuming APIS with python and pandas programmatically
- Programmatic requests


```python
import pandas as pd
import requests

REQ="https://api.figshare.com/v2/projects?"

parameters = {
    "page_size": 100,
    "order": "published_date",
    "order_direction": "desc"
}

# Here we store the response from the http API call
response = requests.get(REQ, params=parameters)

data = response.json()
modified_dates= [item['modified_date'] for item in data]
titles=[item['title'] for item in data]

df= pd.DataFrame({
    'title': titles,
    'modified date': modified_dates
})

df.head(20)
```
